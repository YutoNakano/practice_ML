# 行列演算の基礎

```python
import numpy as np
```

```python
#ベクトルの定義
x = np.array([[1],[2],[3]])
x
```

```python
#行列の定義
X = np.array([[1,2],[3,4]])
print(X)
```

```python
#転置
Xt = X.T
Xt
```

```python
# 逆行列
# linear algebra: 線形代数
X_inv = np.linalg.inv(X)
print(X_inv)
```

```python
#行列積
XX_inv = np.dot(X,X_inv)
print(XX_inv)
```

# 演習問題

```python
#データの定義
X = np.array([
    [1,2,3],
    [1,2,5],
    [1,3,4],
    [1,5,9]
])
print(X)
```

```python
#yの定義
y = np.array([
    [1],
    [5],
    [6],
    [8],
])
print(y)
```

```python
#Step1 行列積：dot
XtX = np.dot(X.T,X)
print(XtX)
```

```python
# Step2 
XtX_inv = np.linalg.inv(XtX)
print(XtX_inv)
```

```python
#Step3
Xty = np.dot(X.T,y)
print(Xty)
```

```python
#Step4
w = np.dot(XtX_inv,Xty)
print(w)
```

# 分布の確認

# 外れ値除去( 3σ法 )

```python
col = 'x6'
```

```python
mean = df.mean()
mean
```

```python
mean[col]
```

```python
# 標準偏差（standard deviation)
sigma = df.std()
sigma
```

```python
sigma[col]
```

```python
low = mean[col] - 3 * sigma[col]
low
```

```python
high = mean[col] + 3 * sigma[col]
high
```

```python
df2 = df[(df[col] > low) & (df[col] < high)]
df2
```

```python
len(df)
```

```python
len(df2)
```

```python
# 分布の確認
```

```python
cols = df.columns
```

```python
cols
```

```python
_df = df
for col in cols:
    #3σ法の上限値を設定
    low = mean[col] -3 * sigma[col]
    high = mean[col] + 3 * sigma[col]
    #条件で絞り込み
    _df = _df[(_df[col] > low)&(_df[col]< high)] 
```

```python
#オリジナル
len(df)
```

```python
#3a法適用後
len(_df)
```

サンプルが減る場合の対処法
-  外れ値は取り除く
- 外れ値を平均もしくは中央値で埋める
- 主成分分析等を使って、潜在変数に変換した後に3σ法を適用　⇦　高度
